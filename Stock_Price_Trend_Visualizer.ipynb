```python
# Install required libraries (Colab)
!pip install yfinance --quiet
!pip install ipywidgets --quiet


import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
from ipywidgets import interact, Dropdown

# Define stock options
stock_options = {
    'Reliance': 'RELIANCE.NS',
    'TCS': 'TCS.NS',
    'Infosys': 'INFY.NS',
    'ICICI Bank': 'ICICIBANK.NS',
    'HDFC Bank': 'HDFCBANK.NS',
    'SBIN': 'SBIN.NS',
    'Apple': 'AAPL',
    'Tesla': 'TSLA',
    'Google': 'GOOGL',
    'Amazon': 'AMZN'
}
def visualize_stock(selected_symbol):
    # Download stock data
    data = yf.download(selected_symbol, start='2022-01-01', end='2024-12-31')

     # Show top 5 rows
    display(data.head())

    # --- 1. Plot: Close Price ---
    plt.figure(figsize=(14, 6))
    plt.plot(data['Close'], label='Close Price', color='blue')
    plt.title(f"{selected_symbol} - Closing Price")
    plt.xlabel("Date")
    plt.ylabel("Price")
    plt.legend()
    plt.grid(True)
    plt.show()

    # --- 2. Plot: 30-Day Moving Average ---
    data['30MA'] = data['Close'].rolling(window=30).mean()
    plt.figure(figsize=(14, 6))
    plt.plot(data['Close'], label='Close Price', color='blue')
    plt.plot(data['30MA'], label='30-Day MA', color='orange')
    plt.title(f"{selected_symbol} - 30-Day Moving Average")
    plt.xlabel("Date")
    plt.ylabel("Price")
    plt.legend()
    plt.grid(True)
    plt.show()

# Launch dropdown
interact(visualize_stock, selected_symbol=Dropdown(options=stock_options, description="Select Stock:"));



```
